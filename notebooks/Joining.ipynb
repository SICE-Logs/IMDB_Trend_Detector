```python
import pandas as pd
import yaml
import csv

path_month = ['2023-10', '2023-11', '2023-12', '2024-01', '2024-02', '2024-03', '2024-04', '2024-05', '2024-06','2024-07', '2024-08', '2024-09', '2024-10', '2024-11']
path_date = pd.date_range(start='2023-10-01', end='2024-11-30').strftime('%Y-%m-%d').tolist()
path_time = ['05-30-00']  # Add all possible times here

wholedf = pd.DataFrame()
for i in path_month:
    for j in path_date:
        for t in path_time:
            file_path = f'D:\\Projects\\Python NBs and Files\\MINI PROJECT_2\\{i}\\{j}_{t}.yaml'
            try:
                with open(file_path, 'r') as yaml_file:
                    yaml_data = yaml.safe_load(yaml_file)
                    if yaml_data:
                        headers = list(yaml_data[0].keys())
                        rows = [list(item.values()) for item in yaml_data]
                    else:
                        headers = []
                        rows = []
                    with open('combined_output_cs.csv', 'a', newline='') as csv_file:  # Use 'a' to append
                        writer = csv.writer(csv_file)
                        if csv_file.tell() == 0:  # Write header only once
                            writer.writerow(headers)
                        writer.writerows(rows)
            except FileNotFoundError:
                continue

```
